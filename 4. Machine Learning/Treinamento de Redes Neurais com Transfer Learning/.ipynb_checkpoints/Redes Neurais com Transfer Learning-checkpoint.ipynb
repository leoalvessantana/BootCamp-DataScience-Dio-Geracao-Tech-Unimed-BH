# Classificação de imagens usando CNN e Transfer Learning



Bibliotecas

```python
import pandas as pd
import numpy as np

import tensorflow as tf
import os, glob

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from tensorflow import keras
from keras.utils.vis_utils import plot_model
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## Carregando e extraindo os dados


```python
# Diretório com as imagens de cagos para treinamento 
train_cat_dir = os.path.join('Dataset\Cat')
file_cats = os.listdir(train_cat_dir)

# Diretório com as imagens de cachorros para treinamento 
train_dog_dir = os.path.join('Dataset\Dog')
file_dogs = os.listdir(train_dog_dir)
```

 Quantidade de imagens em cada diretório

```python
print('Total de imagens de gatos para treinamento:', len(file_cats))
print('Total de imagens de cachorros para treinamento:', len(file_dogs))
```

Analisando as imagens (escolhemos algumas figuras de forma aleatoria)

```python
plt.figure(figsize=(15,7))
for i, c in enumerate(np.random.randint(0,len(file_cats),5), start=1):
    im = mpimg.imread('./Dataset/Cat/'+file_cats[c])
    plt.subplot(2,5,i)
    plt.imshow(im)
    
for i, c in enumerate(np.random.randint(0,len(file_dogs),5), start=6):
    im = mpimg.imread('./Dataset/Dog/'+file_dogs[c])
    plt.subplot(2,5,i)
    plt.imshow(im)
```

## Criando Modelo CNN

```python
from tensorflow.keras.layers import *
from functools import partial

WIDTH, HEIGHT = 150, 150
batch_size = 32

DefaultConv2D = partial(keras.layers.Conv2D, kernel_size=3, activation='relu', padding="SAME")

model = keras.models.Sequential()

model.add(DefaultConv2D(filters=32, kernel_size=5, input_shape=(WIDTH, HEIGHT, 3)))
model.add(MaxPooling2D(pool_size=2))

model.add(DefaultConv2D(filters=64))
model.add(DefaultConv2D(filters=64))
model.add(MaxPooling2D(pool_size=2))

model.add(DefaultConv2D(filters=128))
model.add(DefaultConv2D(filters=128))
model.add(MaxPooling2D(pool_size=2))

model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
```

## Configurar as especificações para o treinamento do modelo

```python
model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(lr=0.001),
              metrics=['accuracy'])
```

## Pré-processamento de dados

```python
# Mudar a escala de todas as imagens para 1/255
train_datagen = ImageDataGenerator(rescale=1/255)

# Imagens de treinamento de fluxo em lotes de 128 usando o train_datagen generator
train_generator = train_datagen.flow_from_directory(
        'Dataset\\',  # Diretório de imagens de treinamento
        target_size=(WIDTH, HEIGHT),  # Todas as imagens em 300x300
        batch_size=batch_size,
        # estamos usando o binary_crossentropy loss, então precisamos de rótulos binários
        class_mode='binary')
```

## Treinando Modelo

```python
history = model.fit_generator(
      train_generator,
      steps_per_epoch=8,  
      epochs=15,
      verbose=2)

model.save('cat_dog_model.h5')
```

## Mostrar como a precisão evolui ao longo dos períodos de treinamento

```python
plt.plot(history.history['accuracy'], c='c', lw=2)
plt.grid(True)
plt.title("Precisão de treinamento com epochs\n", fontsize=18)
plt.xlabel("Treinamento epochs", fontsize=15)
plt.ylabel("Accuracy de treinamento", fontsize=15)
plt.show()
```

## Predição do nosso modelo

```python
import numpy as np
from keras.preprocessing import image

path_img = 'D:\\Teste\\'

for arquivo in glob.glob(os.path.join(path_img, '*')):
    img = image.load_img(arquivo, target_size=(300, 300))
    plt.imshow(img)
    plt.show()
    
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    images = np.vstack([x])
    
    classes = model.predict(images, batch_size=10)
    
    if classes[0]>0.5:
        print(arquivo.split('\\')[-1] + " shows a human image")
    else:
        print(arquivo.split('\\')[-1] + " shows a horse image")
```
